```python
def imageData(file):
    return Dataset(file, 'r').variables['image_data']
```

```python
indir = "/Users/rmoraski/Library/Mobile \
Documents/com~apple~CloudDocs/School/Electrons/Mathematica/NISTView-\
ish/"
imSuffix = "_ix.sempa"
mxSuffix = "_mx.sempa"
mySuffix = "_my.sempa"

name = "191001_Fe_whisker/Fe_whisker"
run = 2

file = indir + name + f'{run:0>3}'
```

```python
file
```

```python
ncDataset = Dataset(file + imSuffix, 'r')
attributes = ncDataset.ncattrs()
dimensions = [dim for dim in ncDataset.dimensions]
variables = [var for var in ncDataset.variables]
imData = ncDataset.variables['image_data']
```

```python
print('Attributes: ', attributes)
print('Dimensions: ', dimensions)
print('Variables : ', variables)
imData
```

```python
ddt = np.transpose(imageData(file + imSuffix))
```

```python
nddt = normalize(ddt)
print(['Min of intensity: ', np.min(ddt), ', max: ', np.max(ddt), ', range: ', np.max(ddt) - np.min(ddt)])
plt.imshow(ddt, cmap='gray', vmin=np.min(ddt), vmax=np.max(ddt));
plt.title('Raw');
plt.show();

print(['Normalized min of intensity: ', np.min(nddt), ', max: ', np.max(nddt), ', range: ', np.max(nddt) - np.min(nddt)])
```

```python
plt.figure(figsize=(12, 6));

plt.subplot(1, 4, 1);
plt.imshow(ddt, cmap='gray');
plt.title('Raw');

plt.subplot(1, 4, 2);
plt.imshow(ddt, cmap='gray', norm=Normalize());
plt.title('im Normalize');

plt.subplot(1, 4, 3);
plt.imshow(ddt, cmap='gray', vmin=np.min(ddt), vmax=np.max(ddt));
plt.title('im vmin, vmax');

plt.subplot(1, 4, 4);
plt.imshow(nddt, cmap='gray');
plt.title('Normalized');
```
